```
#import numpy and pyplot
import numpy as np
import matplotlib.pyplot as plt


#import sample data using np.loadtxt
sample_data = np.loadtxt("/content/drive/My Drive/Sample Data.txt")
```

Plot 2 histograms on the same plot below. Assume the first column of the data is data A and second row is data B

```
#Save the first column and second column seperately
sample_A = sample_data[:,0]
sample_B = sample_data[:,1]

#plot 2 histograms on the same plot, remember to include alpha number and color in the plotting code

plt.hist(sample_A, bins = 87, alpha = 0.3, color = 'red', label = "Sample A")
plt.hist(sample_B, bins = 87, alpha = 0.3, color = 'green' , label = "Sample B")

#add title, x axis lable, y axis lable in the plot

plt.xlabel("Data Number")
plt.ylabel("# of occurances of number")

#Optional: try to add legend to the plot

plt.legend()

```

Optional: calculate and print out the mean and standard diviation of data A and B and print it out using np.mean and np.std. Note you will have to do this in lab!

```
# mean

print("Mean of Data Set A is ", np.mean(sample_A))
print("Mean of Data Set B is ",np.mean(sample_B))

# standard Deviation

print("Standard Deviation of Set A is ", np.std(sample_A))
print("Standard Deviation of Set B is ", np.std(sample_B))
```
