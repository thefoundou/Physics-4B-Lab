# Exercise 1.2: Using numpy functions

Material for this exercise is covered in [Tutorial 1](https://drive.google.com/file/d/1MaqFfsu1FyTT7V2Yv8toP9nbGc1l8ith/view?usp=sharing) and [Youtube video](https://youtu.be/npjSBqiJGnY) 2

Start by importing numpy below

```
# import module:
import numpy as np
```

Now use the `np.sin` function to calculate 

`sin_value` = $\sin( k * t )$

Remember that you defined your variables above


```
# Create sin_value

sin_value = np.sin(k*t)
```

Now use the `np.exp` function to find the exponential value of (-a*t)


`exponential_value` = $e^{-at}$



```
# Define exponential_value

exponential_value = np.exp(-a*t)
```

Multiply the numbers to create

$exponential\_sin = \sin (k t) e^{-at}$

```
exponential_sin = sin_value * exponential_value
```

```
print(exponential_sin)
```
